```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取CSV文件
data = pd.read_csv('./TaskIntervalEnv/prime_seed_0_0429_02-56-06/PPO_TaskIntervalEnv_log.csv')

# 提取第二列和第三列数据作为横坐标和纵坐标
x = data.iloc[:, 1]  # 第二列
y = data.iloc[:, 2]  # 第三列

plt.figure(figsize=(30, 6))

# 绘制折线图
plt.plot(x, y, marker='none', linestyle='-')

# 添加标题和标签
plt.title('Reward vs Timestep')
plt.xlabel('Timestep')
plt.ylabel('Reward')

# 显示图形
plt.show()
```
