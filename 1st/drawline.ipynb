```python
import pandas as pd
import matplotlib.pyplot as plt

# CSV
data = pd.read_csv('./TaskIntervalEnv/prime_seed_0_0429_02-56-06/PPO_TaskIntervalEnv_log.csv')

# 
x = data.iloc[:, 1]  # 
y = data.iloc[:, 2]  # 

plt.figure(figsize=(30, 6))

# 
plt.plot(x, y, marker='none', linestyle='-')

# 
plt.title('Reward vs Timestep')
plt.xlabel('Timestep')
plt.ylabel('Reward')

# 
plt.show()
```
